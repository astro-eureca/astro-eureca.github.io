```python

# Path to the directory containing input files
input_dir = '/Volumes/Expansion/EE_1567/'
output_dir = input_dir
```

```python
run_detector1 = True # set to False if you have already run this pipeline. It takes a while to run this on all the data
if run_detector1: 
    file_uncal = glob(input_dir+'*uncal.fits')
    for file in file_uncal:
        detect1 = Detector1Pipeline()
        result = detect1.call(file,
                          steps={'jump': {'rejection_threshold': 3,
                                          'expand_large_events': True, # to take care of snowballs
                                          'sat_required_snowball': False, # workaround?
                                          'min_jump_area': 10, # avoid a bunch of false snowball detections
                                          'after_jump_flag_dn1': 1000, # how big of a jump triggers the flag
                                          'after_jump_flag_time1': 10, # just skip one group (this is in seconds)
                                          'maximum_cores': 'half'
                                          },
                                 'ramp_fit': {'maximum_cores': 'half'}
                                },
                          save_results=True,
                          output_dir=output_dir,
                          input_dir=input_dir)
    
```

```python
# Use the assocations to run calspec2. They will contain the background observations to combine
# and subtract

run_spec2 = True
if run_spec2:
    asn_files = glob(input_dir + 'jw01567-o002*spec2*json')
    for asn in asn_files:
        print(asn)
        spec2 = Spec2Pipeline()
        result = spec2.call(asn,
                          save_results=True,
                          output_dir=output_dir)
```

```python
run_spec3 = True
if run_spec3:
    asn_files = glob(input_dir + 'jw01567-o002*spec3*json')
    for asn in asn_files:
        print('asn',asn)
        spec3 = Spec3Pipeline()
        result = spec3.call(asn,
                        steps={'outlier_detection': {'skip': False,
                                                  'kernel_size': '5 5', # to take care of snowballs
                                                  'threshold_percent': 97.5, # workaround?
                                                  'ifu_second_check': True, # avoid a bunch of false snowball detection
                                          },
                              'cube_build': {'coord_system': 'ifualign'},
                            
                                },
                          save_results=True,
                          output_dir=output_dir)
        
```

```python
# IF there are spikes other artifacts in s3d files - try to clean up the cal files. 

# This is something I have added and is not standard. I have found that working with NIRSpec IFU data some some of the reference files do 
# not have the DQ flags set to DO_NOT_USE, but other DQ flags indicate the data is not useable. By try and error I found these flags useful
# to set to DO_NOT_USE so the spec3 pipeline (cube_build_does not used them )
# The NIRSPec team is currently updating reference files and flagging DQ values in the reference files appropriately. So eventully this 
# will not be neded. 
# You could turn off some of these flagging and  compare the *_cal.fits to the  *_cal_mod.fits and double check that the  flagging
# does flag data you do not want 

DO_NOT_USE = datamodels.dqflags.pixel["DO_NOT_USE"]
NO_SAT_CHECK = datamodels.dqflags.pixel["NO_SAT_CHECK"]
UNRELIABLE_FLAT = datamodels.dqflags.pixel["UNRELIABLE_FLAT"]
OTHER_BAD_PIXEL = datamodels.dqflags.pixel['OTHER_BAD_PIXEL']
MSA_FAILED_OPEN = datamodels.dqflags.pixel['MSA_FAILED_OPEN']
outstr_dq = '_cal_dqfix.fits'

clean_data = True # turn to True to flag additional pixels - False will not set additional DQ flags
if clean_data:
    
    file_cal = glob(input_dir+'*_cal.fits')
    print(file_cal)
    
    for file in(file_cal):
        print('File', file)
        new_file = file.replace('_cal.fits', outstr_dq)
        print('new file', new_file)
        input_cal = datamodels.IFUImageModel(file)
    
        data1 = input_cal.data
        dq1 = input_cal.dq
        bad1 = np.where(np.bitwise_and(dq1, UNRELIABLE_FLAT).astype(bool))
        nbad1 = len(bad1[0])

        # UNRELIABLE_FLAT CHECK
        # check if these values have DO_NOT_USE SET
        dqtest = dq1[bad1]
        test1 = np.where(np.bitwise_and(dqtest, DO_NOT_USE).astype(bool))
        ntest1 = len(test1[0])
        n1 = nbad1 -ntest1
        
        if n1 > 0:
            print('Number of pixels with UNREIABLE_FLAT but no DO_NOT_USE', n1)
            input_cal.dq[bad1] = np.bitwise_or(input_cal.dq[bad1], datamodels.dqflags.pixel['DO_NOT_USE'])
            input_cal.data[bad1] = np.nan
        # NO_SAT_CHECK
        data2 = input_cal.data
        dq2 = input_cal.dq
        bad2 = np.where(np.bitwise_and(dq2, NO_SAT_CHECK).astype(bool))
        nbad2 = len(bad2[0])
        dqtest2 = dq2[bad2]
        test2 = np.where(np.bitwise_and(dqtest2, DO_NOT_USE).astype(bool))
        ntest2 = len(test2[0])
        
        n2 = nbad2 - ntest2
        if n2 > 0:
            print('Number of pixels with NO_SAT_CHECK but no DO_NOT_USE', n2)
            input_cal.dq[bad2] = np.bitwise_or(input_cal.dq[bad2], datamodels.dqflags.pixel['DO_NOT_USE']) 
            input_cal.data[bad2] = np.nan

        # OTHER_BAD_PIXEL 
        data3 = input_cal.data
        dq3 = input_cal.dq
        bad3 = np.where(np.bitwise_and(dq3, OTHER_BAD_PIXEL).astype(bool))
        nbad3 = len(bad3[0])
        dqtest3 = dq3[bad3]
        test3 = np.where(np.bitwise_and(dqtest3, DO_NOT_USE).astype(bool))
        ntest3 = len(test3[0])
        n3 = nbad3 - ntest3
        if n3 > 0:
            print('Number of pixels with OTHER_BAD_PIXEL but not DO_NOT_USE', n3)
            input_cal.dq[bad3] = np.bitwise_or(input_cal.dq[bad3], datamodels.dqflags.pixel['DO_NOT_USE'])
            input_cal.data[bad3] = np.nan
        # MSA FAILED OPEN
        data4 = input_cal.data
        dq4 = input_cal.dq
        bad4 = np.where(np.bitwise_and(dq4, MSA_FAILED_OPEN).astype(bool))
        nbad4 = len(bad4[0])
        dqtest4 = dq2[bad4]
        test4 = np.where(np.bitwise_and(dqtest4, DO_NOT_USE).astype(bool))
        ntest4 = len(test4[0])

        n4 = nbad4 - ntest4
        if n4 > 0:
            #print('Number of pixels with MSA Failed Open but no DO_NOT_USE', n4)
            #input_cal.dq[bad4] = np.bitwise_or(input_cal.dq[bad4], datamodels.dqflags.pixel['DO_NOT_USE'])
            #input_cal.data[bad4] = np.nan
    
       
        
        input_cal.save(new_file)
print('Done first clean') 
          
```
